```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import yfinance as yf
import datetime as dt
import plotly.graph_objects as go
from PIL import Image
import os
import glob
```

```python
# List of stock
tickers = ['ICICIBANK.NS']
for stock in tickers:
    # Ticker = "ICICIBANK.NS"
    start_date = "2022-01-01"
    end_date = "2024-12-31"
    download_data = yf.download(tickers= stock, start=start_date, end=end_date)
    data = pd.DataFrame(download_data['Close'])
    # return_final_output(data,stock)
    # plt.savefig(f"../top_50/{stock}.png")
    # save_outcome_image(stock)
```

```python
data['mean'] = data['Close'].rolling(30).mean()
data['std'] = data['Close'].rolling(30).std()
data['standarized_value'] = (data['Close'] - data['mean'])/data['std']

fig, ax = plt.subplots(4,figsize=(12,8))
ax[0].plot(data['Close'])
# ax[1].plot(data['mean'])
# ax[2].plot(data['std'])
ax[3].plot(data['standarized_value'])
ax[3].grid(True)
plt.show()
```

```python
from datetime import datetime
import yfinance as yf
import pandas as pd
```

```python
class LoadData:
    """This class will load data from both yfinance and the excelbook"""
    def __init__(self, stock):
        self.stock = stock
        self.start_date = '2014-01-01'
        self.end_date = datetime.today().strftime('%Y-%m-%d')

    def download_yfinace_data(self, stock):
        yf_data = yf.download(self.stock, start=self.start_date, end = self.end_date)
        yf_data = yf_data.resample('D').ffill()
        yf_data = pd.DataFrame(yf_data['Close'])
        return yf_data

    def load_p_n_l_excel_data(self, stock):
        data_p_n_l = pd.read_excel(f"../../data/{stock}.xlsx", skiprows=15, sheet_name = "Data Sheet", nrows=15)
        data_p_n_l = data_p_n_l.T

        # update header
        new_header = data_p_n_l.iloc[0] #grab the first row for the header
        data_p_n_l = data_p_n_l[1:] #take the data less the header row
        data_p_n_l.columns = new_header #set the header row as the df header
        data_p_n_l['Date'] = pd.to_datetime(data_p_n_l.index)
        data_p_n_l = data_p_n_l.set_index(data_p_n_l['Date'])
        # data_p_n_l = data_p_n_l.drop(columns=["Date"])
        data_p_n_l = data_p_n_l.dropna(axis=1)
        return data_p_n_l

    def load_quar_excel_data(self, stock):
        data_quarter = pd.read_excel("../../data/Infosys.xlsx", skiprows=40, sheet_name = "Data Sheet", nrows=10)
        data_p_n_l = data_quarter.T

        # update header
        new_header = data_p_n_l.iloc[0] #grab the first row for the header
        data_p_n_l = data_p_n_l[1:] #take the data less the header row
        data_p_n_l.columns = new_header #set the header row as the df header
        data_p_n_l['Date'] = pd.to_datetime(data_p_n_l.index)
        data_p_n_l = data_p_n_l.set_index(data_p_n_l['Date'])
        # data_p_n_l = data_p_n_l.drop(columns=["Date"])
        data_p_n_l = data_p_n_l.dropna(axis=1)
        return data_p_n_l
```

```python
class merge_price_data:
    """merger price data with other dataframes calculated like pnl and qyuar"""
    def __init__(self, yf_ticker, excel_file_data):
        self.yf_ticker = yf_ticker
        self.excel_file_data = excel_file_data

    def merge_pnl_with_close(self, yf_ticker,excel_file_data):
        data = LoadData(yf_ticker)
        yf_data = data.download_yfinace_data(yf_ticker)

        data_p_n_l= data.load_p_n_l_excel_data(excel_file_data)
        
        price = yf_data.loc[data_p_n_l['Date']]['Close']
        # price.values
        data_p_n_l['Close'] = price.values
        return data_p_n_l

    def merge_quar_with_close(self, yf_ticker,excel_file_data):
        data = LoadData(yf_ticker)
        yf_data = data.download_yfinace_data(yf_ticker)

        data_p_n_l= data.load_quar_excel_data(excel_file_data)
        
        price = yf_data.loc[data_p_n_l['Date']]['Close']
        # price.values
        data_p_n_l['Close'] = price.values
        return data_p_n_l
```
